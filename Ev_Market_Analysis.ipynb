**Sales of EV State Wise**

```python



plt.figure(figsize=(15, 9))
sns.barplot(x='Total Number of Invoice/Sales', y='State/UT', data=df.sort_values('Total Number of Invoice/Sales',ascending=False),palette='Set1')
plt.xlabel('Total Number of Invoice/Sales')
plt.ylabel('Sales/UT')
plt.title('State/UT vs Total Number of Invoice/Sales')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

**State with high Sales of EV**

```python

df_sorted = df.sort_values('Total Number of Invoice/Sales', ascending=False)
df_sorted.head()
```

```python


plt.figure(figsize=(15, 9))
sns.barplot(x='Total Number of Invoice/Sales', y='State/UT', data=df_sorted, palette='viridis')  # Choose a different palette
plt.xlabel('Total Number of Invoice/Sales')
plt.ylabel('State/UT')
plt.title('State/UT vs Total Number of Invoice/Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
