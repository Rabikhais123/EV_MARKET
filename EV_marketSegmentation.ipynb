**Data Preprocessing**

```python
selected_columns = ['Age', 'Profession', 'Marrital Status', 'Education', 'No of Dependents',
                    'Personal loan', 'House Loan', 'Wife Working', 'Salary', 'Wife Salary', 'Total Salary']
```

```python

from sklearn.preprocessing import StandardScaler, LabelEncoder
label_encoder = LabelEncoder()
for col in ['Profession', 'Marrital Status', 'Education', 'Personal loan', 'House Loan', 'Wife Working']:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df[selected_columns])
```

```python
#  Using the Elbow Method to find the optimal number of clusters
wcss = []  # Within-cluster sum of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(data_scaled)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method to Determine Optimal Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_scores = []
k_values = range(2, 11)  # Testing k values from 2 to 10

for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    clusters = kmeans.fit_predict(data_scaled)
    silhouette_avg = silhouette_score(data_scaled, clusters)
    silhouette_scores.append(silhouette_avg)

# Step 4: Plot Silhouette Scores
plt.figure(figsize=(10, 6))
plt.plot(k_values, silhouette_scores, marker='o', linestyle='--')
plt.title('Silhouette Scores for Different Number of Clusters')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

```python
optimal_k = k_values[silhouette_scores.index(max(silhouette_scores))]
print(f'Optimal number of clusters based on Silhouette Method: {optimal_k}')
```

```python
#Apply K-Means clustering with the optimal number of clusters
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(data_scaled)
```

```python
df['Cluster'] = clusters
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Age'], y=df['Total Salary'], hue=df['Cluster'], palette='viridis', s=100)
plt.title('Clusters of Customers based on Age and Total Salary')
plt.xlabel('Age')
plt.ylabel('Total Salary')
plt.grid(True)
plt.show()
```

Cluster 0 (purple) seems to consist of younger individuals with lower salaries.

Cluster 1 (blue) includes slightly older individuals with varying salaries.

Cluster 2 (dark green) seems to show a mix of ages but predominantly higher salaries.

Cluster 3 (light green) has individuals mostly around middle age with high salaries.

Cluster 4 (yellow) consists of older individuals, potentially indicating a demographic with the highest earnings.
